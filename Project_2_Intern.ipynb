```python
# Create a copy of the dataset
df_clean = df.copy()

# Convert date column to datetime format
df_clean['date'] = pd.to_datetime(df_clean['date'])

# Remove duplicate rows
df_clean = df_clean.drop_duplicates()

print("Rows after cleaning:", len(df_clean))
```

```python
df_sorted = df_clean.sort_values(['song', 'date'])

df_sorted['prev_song'] = df_sorted['song'].shift(1)

df_sorted['is_new_streak'] = (
    df_sorted['song'] != df_sorted['prev_song']
)

reentry = df_sorted.groupby(['song', 'artist'])['is_new_streak'].sum().reset_index()

reentry = reentry.rename(columns={'is_new_streak': 'reentry_count'})

reentry.sort_values('reentry_count', ascending=False).head(20)
```

```python
reentry.describe()
```

```python
df_re = df_clean.copy()

# Sort properly for time analysis
df_re = df_re.sort_values(['song', 'date'])

# Check if song was present previous day
df_re['prev_presence'] = df_re.groupby('song')['position'].shift(1)

# If previous is NaN → first entry
df_re['is_first_entry'] = df_re['prev_presence'].isna()

# Re-entry happens when song appears again after missing days
df_re['is_reentry'] = (
    (df_re['is_first_entry'] == False) &
    (df_re['prev_presence'].isna())
)

# Correct method: detect gaps in presence
df_re['was_present_before'] = df_re.groupby('song')['date'].diff().dt.days > 1

df_re['reentry_flag'] = (
    df_re['was_present_before'].fillna(False)
)

reentry_final = df_re.groupby(['song','artist'])['reentry_flag'].sum().reset_index()

reentry_final.rename(columns={'reentry_flag':'reentry_count'}, inplace=True)

reentry_final.sort_values('reentry_count', ascending=False).head(20)
```

```python
momentum_df = df_clean.copy()

# Merge re-entry data
momentum_df = momentum_df.merge(
    reentry_final,
    on=['song','artist'],
    how='left'
)

momentum_df['reentry_count'] = momentum_df['reentry_count'].fillna(0)

# Create song-level metrics
song_metrics = momentum_df.groupby(['song','artist']).agg(
    avg_popularity=('popularity','mean'),
    total_days=('date','count'),
    reentry_count=('reentry_count','max')
).reset_index()

# Normalize simple scores
song_metrics['momentum_score'] = (
    song_metrics['avg_popularity'] * 0.5 +
    song_metrics['total_days'] * 0.3 +
    song_metrics['reentry_count'] * 10
)

song_metrics.sort_values('momentum_score', ascending=False).head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

song_metrics_scaled = song_metrics.copy()

song_metrics_scaled[['avg_popularity','total_days','reentry_count']] = scaler.fit_transform(
    song_metrics_scaled[['avg_popularity','total_days','reentry_count']]
)

song_metrics_scaled['momentum_score_final'] = (
    song_metrics_scaled['avg_popularity'] * 0.5 +
    song_metrics_scaled['total_days'] * 0.3 +
    song_metrics_scaled['reentry_count'] * 0.2
)

song_metrics_scaled.sort_values('momentum_score_final', ascending=False).head(20)
```

```python
#Categorize Momentum Levels
def momentum_level(score):
    if score >= 0.75:
        return "High"
    elif score >= 0.40:
        return "Medium"
    else:
        return "Low"

song_metrics_scaled['Momentum_Level'] = song_metrics_scaled['momentum_score_final'].apply(momentum_level)

song_metrics_scaled[['momentum_score_final', 'Momentum_Level']].head()
```

```python
#Rank Recovery Speed Analysis
#Sort The Dataset
rank_df = df_clean.copy()

rank_df['date'] = pd.to_datetime(rank_df['date'], dayfirst=True)

rank_df = rank_df.sort_values(['song', 'date'])
```

```python
#Previous Rank
rank_df['previous_position'] = rank_df.groupby('song')['position'].shift(1)

rank_df[['song', 'date', 'position', 'previous_position']].head()
```

```python
#Calculate Rank Recovery
rank_df['rank_recovery'] = rank_df['previous_position'] - rank_df['position']

rank_df[['song', 'position', 'previous_position', 'rank_recovery']].head()
```

```python
#Average Recovery By Song
recovery_score = (
    rank_df.groupby('song')['rank_recovery']
    .mean()
    .reset_index()
)

recovery_score = recovery_score.sort_values(
    by='rank_recovery',
    ascending=False
)

recovery_score.head(15)
```

```python
#Visualization
fig = px.bar(
    recovery_score.head(15),
    x='rank_recovery',
    y='song',
    orientation='h',
    title='Top 15 Songs by Average Rank Recovery'
)

fig.show()
```
